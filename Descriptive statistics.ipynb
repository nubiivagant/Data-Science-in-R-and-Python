```python
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis
```

```python
data = [12, 15, 14, 10, 18, 20, 22, 15, 14, 13, 17, 19, 21]
```

```python
series = pd.Series(data)
```

```python
# Descriptive statistics
mean_val = series.mean()
median_val = series.median()
std_val = series.std()
variance_val = series.var()
skewness_val = skew(series)
kurtosis_val = kurtosis(series)
```

```python
print("Descriptive Statistical Analysis:")
print(f"Mean      : {mean_val:.2f}")
print(f"Median    : {median_val:.2f}")
print(f"Std Dev   : {std_val:.2f}")
print(f"Variance  : {variance_val:.2f}")
print(f"Skewness  : {skewness_val:.2f}")
print(f"Kurtosis  : {kurtosis_val:.2f}")
```

Using random data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
data = np.random.normal(loc=50, scale=10, size=1000)
```

```python
series = pd.Series(data)
```

```python
# Descriptive statistics
mean_val = series.mean()
median_val = series.median()
std_val = series.std()
variance_val = series.var()
skewness_val = skew(series)
kurtosis_val = kurtosis(series)
```

```python
# Display results
print("Descriptive Statistical Analysis:")
print(f"Mean      : {mean_val:.2f}")
print(f"Median    : {median_val:.2f}")
print(f"Std Dev   : {std_val:.2f}")
print(f"Variance  : {variance_val:.2f}")
print(f"Skewness  : {skewness_val:.2f}")
print(f"Kurtosis  : {kurtosis_val:.2f}")
```

```python
plt.figure(figsize=(14,6))

# Histogram + KDE
plt.subplot(1,2,1)
sns.histplot(series, kde=True, bins=30, color="skyblue")
plt.axvline(mean_val, color='red', linestyle='--', label=f"Mean = {mean_val:.2f}")
plt.axvline(median_val, color='green', linestyle='--', label=f"Median = {median_val:.2f}")
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
```

```python
# Boxplot
plt.subplot(1,2,2)
sns.boxplot(x=series, color="pink")
plt.title("Boxplot")
plt.xlabel("Value")

plt.tight_layout()
plt.show()
```
